```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import statistics
from scipy.stats import spearmanr, kendalltau, pearsonr, boxcox
%matplotlib inline
```

```python
df = pd.read_csv("/root/ml_mate_exoplanet/Data/PSCompPars_2021.05.07_10.28.24.csv", sep=",")
#'hostname', 'sy_pnum', 'pl_bmasse', 'pl_bmassj'
var_df = df[['sy_pnum', 'st_mass','hostname', 'sy_pnum', 'pl_bmasse', 'pl_bmassj']]
var_df.head()
```

```python
pl_num = []


for g in var_df.sy_pnum.tolist():
    if g == 2:
        ind = var_df.sy_pnum.tolist().index(g)
        pl_num.append(ind)
        pl_num.append(ind + 1)
```

```python
pl_nam_sy = np.array(var_df.hostname.tolist())
name = []
z = []
name_2 = []
for i in pl_nam_sy:
    if i not in name:
        name.append(i)

for elem in name:
    indic = np.where(pl_nam_sy==elem)
    m = np.array(var_df.pl_bmasse.tolist())[indic]
    add = 0
    for g in m:
        add+=g
    z.append(add)

z_2 = []

for a in name:
    indic_2 = np.where(pl_nam_sy==a)
    m_2 = np.array(var_df.st_mass.tolist())[indic_2]

    z_2.append(m_2[0])


```

```python
print(len(name), len(z_2))

plt.scatter(z, z_2)
plt.show()
```
